### Step 9. Which was the most-ordered item? 

```python
df2 = df.groupby("item_name").count().sort_values("quantity",ascending=False)
df2.head(20)
first_row = df2.iloc[0]
print(f"the most ordered item was {first_row.name}")
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
df2 = df.groupby("item_name").sum().sort_values(["quantity"],ascending=False)
df2.head(20)
first_row2 = df2.iloc[0]
print(f" {first_row.name} was ordered {first_row.quantity} times")
```

### Step 11. What was the most ordered item in the choice_description column?

```python
df3 = df.groupby("choice_description").sum().sort_values(["quantity"],ascending=False)
df3.head(20)
first_row3 = df3.iloc[0]
print(f" {first_row3.name} was ordered {first_row3.quantity} times")
```

### Step 12. How many items were orderd in total?

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
df.item_price.info
```

#### Step 13.b. Change the type of item price

```python
df["item_price"] = df['item_price'].str.replace('$', '').astype(float)
```

#### Step 13.c. Check the item price type

```python
df.item_price.info
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (df["item_price"] * df["quantity"]).sum()
print(f" the revenue was {revenue} $")
```

### Step 15. How many orders were made in the period?

```python
df.order_id.nunique()
```

### Step 16. What is the average revenue amount per order?

```python
average = (df["item_price"] * df["quantity"]).sum() / df.order_id.nunique()
print(f" the average revenue per order was: {average}")
```
